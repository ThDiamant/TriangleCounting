# Relative error vs samples

```python
dbl_rel_list = [111973, 329503, 669439, 1171624, 2224385]
fb_rel_list = [81186, 400705, 679880, 1454225, 1612010]
slashdot_rel_list = [707381, 1599423, 2165605, 2563325, 2833879]
roadnet_rel_list = [49080, 100065, 155599, 208515, 261799]
as_skitter_rel_list = [5062446, 11445854, 19455296, 25219689, 28769868]
```

```python
def find_rel_error(actual, graphlist):
  errorslist = []
  for size in graphlist:
    relerr = ((actual - size) / actual)*100
    errorslist.append(relerr)
  return errorslist
```

```python
fb_rel = find_rel_error(1612010, fb_rel_list)
slashdot_rel = find_rel_error(3290768, slashdot_rel_list)
dbl_rel = find_rel_error(2224385, dbl_rel_list)
roadnet_rel = find_rel_error(331476, roadnet_rel_list)
as_skitter_rel = find_rel_error(28769868, as_skitter_rel_list)
```

```python
plt.figure(figsize=(16,10))

plt.plot(y, fb_rel, "-o", label='fb graph, |V|=4K, |E|=88K')
plt.plot(y, slashdot_rel, "-o", label='slashdot graph, |V|=77K, |E|=900K')
plt.plot(y, dbl_rel, "-o", label='dbl graph, |V|=300K, |E|=1M')
plt.plot(y, roadnet_rel, "-o", label='roadnet tx graph, |V|=1M, |E|=3M')
plt.plot(y, as_skitter_rel, "-o", label='as-skitter graph, |V|=1.7M, |E|=11M')

plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(xmax=100.0/10))
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100.0/1))

plt.xlabel('Reservoir sample (% of total graph)')
plt.ylabel('Triangle count relative error (%)')
plt.title('Triest base relative error per size of reservoir')
plt.grid()
plt.legend()

```
